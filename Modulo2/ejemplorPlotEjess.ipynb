```python
import matplotlib.pyplot as plt
import numpy as np

x= np.arange(10)
y = np.sin(x*np.pi/180)
```

```python
figure = plt.figure()

plt.plot(x, y,  marker="P", linewidth=4)
plt.xlabel("eje x")
plt.ylabel("eje y")
plt.title("Mi primera gráfica")
plt.savefig("plot1.png")
plt.show()
```

```python
figure = plt.figure()
ax =figure.add_axes([0.12, 0.13, 0.78, 0.85])
ax.plot(x, y)
ax.set_xlabel("eje x")
ax.set_ylabel("eje y")
plt.savefig("plot1_add_axis.png")
plt.show()
```

```python
x1 = np.arange(10)
y1 = np.sin(x1*np.pi)

x2 = np.arange(30)
y2 = np.sin(x2*np.pi)*np.exp(-10)

import matplotlib
params = {'font.size'     : 18,
          'figure.figsize':(9.0, 6.0), # 'figure.figsize':(ncols, nfilas)
          'lines.linewidth': 2.,
          'lines.markersize': 15,
          'lines.linestyle' : "-",
          'lines.marker' : 'o'}
matplotlib.rcParams.update(params)

figure = plt.figure(figsize=(10,8))
ax1 =figure.add_axes([0.12, 0.13, 0.35, 0.82])
ax1.plot(x1, y1)
ax1.set_xlabel("eje x")
ax1.set_ylabel("eje y")

ax2 =figure.add_axes([0.6, 0.13, 0.35, 0.82])
ax2.plot(x2, y2)
ax2.set_xlabel("eje x")
ax2.set_ylabel("eje y")


plt.savefig("plot1_add_axis-multipanel1.png")
plt.show()
```
